```python
import joblib, xgboost as xgb, json, numpy as np

booster = joblib.load("../ml/model.joblib")
with open("../ml/feature_cols.json") as f:
    FEATURE_COLS = json.load(f)

# Fake single-row inference vector (zeros — just testing the pipeline)
import pandas as pd
fake_row = pd.DataFrame([{col: 0.0 for col in FEATURE_COLS}])
dmat = xgb.DMatrix(fake_row, feature_names=FEATURE_COLS)
pred = booster.predict(dmat)
print(f"Cold-load inference: {pred[0]:.4f}")   # any float = success
```
